```python
# Feature Engineering perpare

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = r'C:\Users\Francis Hung\Desktop\Datatrain'
df_train = pd.read_csv(data_path + r'\house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# HW1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y) 
plt.show()  # transform data in regression analysis

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df) 
estimator = LinearRegression()  
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF限制在 800到1800內
df['1stFlrSF'] = df['1stFlrSF'].clip(800,1800)
sns.regplot(x= df['1stFlrSF'],y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# HW2
* 續前題, 去除離群值有兩類方式 : 捨棄離群值(刪除離群的資料) 以及調整離群值,請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF限制在 800到1800內 並捨棄離群值
keep_indexs = (df['1stFlrSF']>800) & (df['1stFlrSF']<1800)   #取閉區間
df = df[keep_indexs] 
train_Y = train_Y[keep_indexs]
sns.regplot(x = df['1stFlrSF'],y=train_Y)
plt.show()


# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

* 可以看出，在資料上下限相同的狀況下，刪除離群資料是有助於cross_val_score的提升，原因可以從OUT[64]和OUT[65]看出，原本的方法是將超過上下界的離群數限制在最大最小值上，這也會使分數下降，因此叫好的做法是直接刪除。
