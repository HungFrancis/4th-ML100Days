# Show other ways to load pictures

1. skimage.io =>  skio.imread("")
   
2. PIL.Image  =>  img2 = Image.open("")

3. OpveCV     => img3 = cv2.imread("")

Method 1

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import skimage.io as skio
img1 = skio.imread(r'C:\Users\Francis Hung\Desktop\Datatrain\FSAE LOGO3-01-01.jpg')
plt.imshow(img1)
plt.show()
```

Method 2 

```python
# using PIL in grabing image from URL
from PIL import Image
img2 = Image.open(r'C:\Users\Francis Hung\Desktop\Datatrain\FSAE LOGO3-01-01.jpg') # 這時候還是 PIL object
img2 = np.array(img2)
plt.imshow(img2)
plt.show()
```

Method 3

```python
import cv2
img3 = cv2.imread(r'C:\Users\Francis Hung\Desktop\Datatrain\FSAE LOGO3-01-01.jpg')
plt.imshow(img3)
plt.show()

img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
plt.show()
```
