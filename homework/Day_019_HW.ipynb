```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

dir_data = r'C:\Users\Francis Hung\Desktop\Datatrain'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
unique_house_type = list(app_train['HOUSETYPE_MODE'].unique())

nrows = len(unique_house_type)
ncols = nrows // 2   # 無條件捨去

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE']== unique_house_type[i], 'AMT_CREDIT'].hist(bins = 50)
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```

```python
plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    sns.distplot(app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]) & (app_train['TARGET'] == 0), 'AMT_CREDIT'], label = "TARGET = 0", hist = True)
    sns.distplot(app_train.loc[(app_train['HOUSETYPE_MODE'] == unique_house_type[i]) & (app_train['TARGET'] == 1), 'AMT_CREDIT'], label = "TARGET = 1", hist = True)
    plt.title(str(unique_house_type[i]))
plt.show()
```
