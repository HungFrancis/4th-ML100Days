# Homework Goal
To use numpy to calculate the corelate Correlation coefficient between to data and draw the scatter plot

```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

# Weak Relative

```python
# Randomly generate two group of number x and y (50 numbers within 0~1000 )
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
# plot the scatter plot！
plt.scatter(x,y)
```

# Strong Relative

```python
x = np.random.randint(0,50,1000)

# let y strong relative to y and plus some
y = x + np.random.randint(0,10,1000)

np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```

# Negative Correlation

```python
x = np.random.randint(0,50,1000)

# let y strong relative to y and plus some
y = -x + np.random.randint(0,10,1000)

np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```
